## The big idea

A quantum computer cannot read a spreadsheet. Before any quantum model can
run, classical numbers must be turned into a quantum state. That translation
step is called **feature encoding**, and it is one of the most important and
least understood design choices in quantum machine learning.

This project implements **three** encoding strategies inside variational
quantum classifiers (VQCs) and compares them against a classical support
vector machine (SVM):

1. **Angle encoding** - each feature becomes a single-qubit rotation angle.
   No entanglement.
2. **Amplitude encoding** - the feature vector is stored directly in the
   amplitudes of a quantum state.
3. **ZZ feature map encoding** - rotations plus entangling gates that mix
   pairs of qubits.

## Hypothesis

> Encoding strategies that incorporate quantum entanglement (the ZZ feature
> map) will produce measurably different decision boundaries than a
> non-entangling encoding (angle encoding), but no quantum encoding is
> expected to surpass a classical RBF-kernel SVM on small benchmark datasets.

A null result against the classical baseline is still a real scientific
finding. The goal is an honest comparison, not a quantum victory.

All experiments in this project run on `AerSimulator`, a high-performance

```python
# Cell 02 - Inspect the AerSimulator backend

from qiskit_aer import AerSimulator

backend = AerSimulator()
print(f"Backend name    : {backend.name}")
print(f"Aer version     : {backend.configuration().backend_version}")
print(f"Max qubits      : {backend.configuration().n_qubits}")
print(f"Max shots       : {backend.configuration().max_shots:,}")
print(f"Simulation type : {backend.configuration().simulator}")
```

---
**Cell 03 - A sanity-check circuit.**
Before trusting any results, confirm the whole stack works end to end by
building and running a tiny two-qubit Bell-state circuit. If you see roughly
50% `00` and 50% `11` (and almost no `01` or `10`), your environment is ready.

```python
# Cell 03 - Build, run, and measure a Bell state

from IPython.display import display
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_distribution

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend), shots=10_000).result()
counts = result.get_counts(qc)
print("Measurement counts:", counts)
display(plot_distribution(counts))
```
